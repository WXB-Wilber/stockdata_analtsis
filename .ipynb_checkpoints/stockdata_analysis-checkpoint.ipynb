### 2.3 清洗异常值

#### 定义一个提取异常值与正常值的函数

```python
def box_plot_outliers(data_ser, box_scale):
    """
    利用箱线图去除异常值
    :param data_ser: 接收 pandas.Series 数据格式
    :param box_scale: 箱线图尺度，默认用 box_plot（scale=3）进行清洗
    :return:
    """
    IQR = box_scale * (data_ser.quantile(0.75) - data_ser.quantile(0.25))
    # 下界值
    val_low = data_ser.quantile(0.25) - IQR*1.5
    # 上界值
    val_up = data_ser.quantile(0.75) + IQR*1.5
    # 异常值
    outlier = data_ser[(data_ser < val_low) | (data_ser > val_up)]
    # 正常值
    normal_value = data_ser[(data_ser > val_low) & (data_ser < val_up)]
    return outlier, normal_value, (val_low, val_up)
```

```python
# 过滤异常值
outlier, normal_value, value = box_plot_outliers(stock_data[:], 3)
outlier = pd.DataFrame(outlier)   # 异常数据
normal_data = pd.DataFrame(normal_value)   # 正常数据
normal_data
```

## （3） 基于时间序列模型分析（方法四）

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

### 3.1 数据预处理

#### 3.1.1 修改名称以方便处理数据

```python
df=pd.read_excel("data_akbilgic.xlsx")
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
df.rename(columns={'TL BASED':'ISE','USD BASED':'ISE.1'},inplace=True)
df.head()
```

```python
plt.figure(figsize=(12,7))
sns.lineplot(x='date',y='TL',data=stock_data)
plt.show()
```

#### 3.1.2 转换数据

```python
df.drop(columns=['SP','ISE.1','DAX','FTSE','NIKKEI','BOVESPA','EU','EM'],inplace=True)
df.date = pd.to_datetime(df.date, dayfirst = True)
df.set_index("date", inplace=True)
df=df.asfreq('b')#Pandas dataframe.asfreq()函数用于将TimeSeries转换为指定频率。此函数可以选择提供填充方法来填充/回填缺失的值。它以指定的频率返回符合新索引的原始数据。
df=df.fillna(method='ffill')
df.head()
```

#### 弱形式平稳性的一个例子是白噪声，因为它遵循以下条件：
#### 1.平均值=0；2.恒定方差；3.滞后之间的自相关为0
#### 平均值几乎等于零；
#### 标准差也几乎等于零
#### 这导致了这样的假设，即这个时间序列是白噪声

#### 

#### 3.2.3 观察数据分布

```python
df.hist()
plt.title("Distribution of values of ISE")
plt.show()
```

#### 直方图证明了均值如何接近于零，并且它遵循高斯分布

#### 要查看时间序列是否具有恒定的方差，我们将数据分割，并检查平均值和方差¶

```python
a = round(len(df)/2)
data1 = df[:a]
data2 = df[a:]
print('前半部分数据的均值: %f'%data1.mean())
print('后半部分数据的均值: %f'%data2.mean())
print('前半部分数据的方差: %f'%data1.var())
print('后半部分数据的方差: %f'%data2.var())
```

#### 可以看出，两个半数据之间的方差是恒定的
####  因此，数据也满足白噪声的第二个假设

### 

### 3.2 时间序列分析

####  3.2.1 自相关性分析

```python
from pandas.plotting import autocorrelation_plot
autocorrelation_plot(df.ISE)
plt.figsize=(100,10)
```

#### 显然，所有自相关值都接近于零，证明时间序列是随机的，并且值之间没有自相关¶

#### 3.2.2 季节性分析

```python
from statsmodels.tsa.seasonal import seasonal_decompose
s_dec_additive = seasonal_decompose(x=df.ISE,model='additive')
s_dec_additive.plot()
plt.figsize=(100,5)
plt.show()
```

### 从上图中我们可以看出：
1.      趋势与观察到的图表相似，没有明显的正或负倾斜
2.      季节序列没有明确的模式，因此图中没有模式
3.      残留物也没有固定模式

### 结论：上述时间序列中没有季节性

### 

### 3.4 时间序列建模

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima_model import ARIMA
```

#### 3.4.1 ADF检验（单位根检验）

```python
def ad_fuller_test(sales):
  results = adfuller(sales)
  labels = ['Test Statistic','p-value','Number of lags used','Number of Observations','Critical Value']
  for values, labels in zip(results,labels):
    print(labels,' : ',str(values))
```

```python
ad_fuller_test(df['ISE'])
```

#### 由于p<0.05，因此否定了零假设，数据是稳定的
1. 测试统计量小于1%、5%、10%的所有值
2. p值为0
3. 由于是白噪声，时间序列的值之间没有自相关，因此滞后=0 
4. 时间序列是静止的，并进一步确认该序列是白噪声

### 3.5 模型定阶

#### 3.5.1 ACF 图（自相关系数图）
#### 绘制从第1个滞后到第0个滞后的图表，与自身的相关性将为1

```python
sm.graphics.tsa.plot_acf(df.ISE,lags=40,zero=False)
plt.title('SP-ACF',size=20)
plt.figsize=(20,5)
plt.show()
```

#### 所有直线都在蓝色显著区域内，表明系数不显著，因此滞后之间没有自相关¶

#### 3.5.2 PACF图（偏自相关系数图）

#### 同样，绘制从第1个滞后到第0个滞后的图表，与自身的相关性将为1

```python
sm.graphics.tsa.plot_pacf(df.ISE,lags=40,zero=False)
plt.show() 
```

#### 与ACF图相同，滞后之间没有部分相关性，这证明了我们对白噪声的假设

#### 如果时间序列是白噪声，则根据定义它是不可预测的（因为它在不同时间的值在统计上是独立的）¶
